```python
import pandas as pd
import glob 
import numpy as np
```

```python
# Data from https://home.treasury.gov/interest-rates-data-csv-archive

df = pd.concat(
    map(pd.read_csv, ['yield-curve-rates-1990-2021.csv', 'daily-treasury-rates2022.csv']), ignore_index=True)
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df['dates_cleaned'] = pd.to_datetime(df['date'])
df
```

```python
df.plot(x="dates_cleaned",y="2_yr")
```

```python
df.plot(x="dates_cleaned",y="10_yr")
```

```python
df.to_csv('treasuryyieldcurve.csv')
```

```python
df.to_json('treasuryyieldcurve.json')
```
